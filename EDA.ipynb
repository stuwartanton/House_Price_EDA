```python
#importing all modules needed for data analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.options.display.max_columns = None # Display all columns of the data frame
sns.set()
```

```python
#importing dataset
df_backup = pd.read_csv('train.csv')
```

```python
df = df_backup.copy()
```

```python
#Percentage of nan values for each feature
features_with_na =[features for features in df.columns if df[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean()*100, 4), ' % missing values')
```

```python
#Finding relationship between Missing values and SalePrice(Dependent Variable)
for feature in features_with_na:
    dfs = df.copy()
    dfs[feature] = np.where(dfs[feature].isnull(), 1, 0)
    dfs.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here, There is clear relationship between missing values and sale price.

```python
#List of all numerical variables
numerical_features = [feature for feature in df.columns if df[feature].dtypes != 'O']

print('Number of numerical Variables: ', len(numerical_features))
```

```python
# Top 5 rows of numerical variables
df[numerical_features].head()
```

```python
#Year Features (Temporal Variables)
year_features = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_features
```

```python
#Exploring content of year features
for feature in year_features:
    print(feature, df[feature].unique())
```

```python
#Relationship between year sold and Sale price
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

```python
#comparing the difference between all the year_features with SalePrice

for feature in year_features:
    if feature!='YrSold':
        dfs=df.copy()
        #difference between year variable and year the house was sold for
        dfs[feature]=dfs['YrSold']-dfs[feature]

        plt.scatter(dfs[feature],dfs['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
#Separating discrete feature from numerical features
discrete_features = [feature for feature in numerical_features if len(df[feature].unique())<25 and feature not in year_features+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_features)))
```

```python
df[discrete_features].head()
```

```python
#Relationship between discrete features and Sale Price
for feature in discrete_features:
    dfs = df.copy()
    dfs.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sales Price')
    plt.title(feature)
    plt.show()
```

There is a clear relationship between discrete variables and Sale Price.

```python
continuous_features = [feature for feature in numerical_features if feature not in discrete_features + year_features + ['Id']]
print('Number of Continous Features: ', len(continuous_features))
```

```python
#Understand the distribution of categorical variables
for feature in continuous_features:
    dfs=df.copy()
    dfs[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
#Logarithmic Transformation

for feature in continuous_features:
    dfs=df.copy()
    if 0 in dfs[feature].unique():
        pass
    else:
        dfs[feature]=np.log(dfs[feature])
        dfs['SalePrice']=np.log(dfs['SalePrice'])
        plt.scatter(dfs[feature],dfs['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        plt.title(feature)
        plt.show()
```

```python
#Outliers

for feature in continuous_features:
    dfs = df.copy()
    if 0 in dfs[feature].unique():
        pass
    else:
        dfs[feature] = np.log(dfs[feature])
        dfs.boxplot(column = feature)
        plt.title(feature)
        plt.show()
        
```

```python
# Categorical Features in DataSet
categorical_features=[feature for feature in df.columns if dfs[feature].dtypes=='O']
print('Number of categorical features: ', len(categorical_features))
```

```python
#Top 5 rows in Categorical features
df[categorical_features].head()
```

```python
#Categorical feature and its categories
for feature in categorical_features:
    print('The feature is {} and number of categories are {}'.format(feature,len(df[feature].unique())))
```

```python
#Relationship between categorical features and Sale Price

for feature in categorical_features:
    dfs=df.copy()
    dfs.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

There exists a relationship between categorical features and Sale Price(Dependent Variable)
